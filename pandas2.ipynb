#Table with movie ratings

```python
data={
    'User ID': [1, 1, 2, 2, 3, 3, 4, 3, 7],
    'Movie Name': ['Interstellar', 'Titanic', 'Inception', 'Avatar', 'Lion King', 'RRR', 'Bahubali', 'After', 'DDLJ'],
    'Rating': [9, 8, 7, 10, 9, 8, 7, 8, 9]
}
#Create DataFrame
df = pd.DataFrame(data)
print(df)
```

```python
#Pivot the table
pivot_table = df.pivot(index='User ID', columns='Movie Name', values='Rating')

#Display the pivoted table
print(pivot_table)
```

```python
#Fill the NAN values with any relevant value
pivot_table.fillna(0)
```
