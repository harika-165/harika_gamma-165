```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load the dataset
df = pd.read_csv("construction_cost_data.csv")  # Using the provided dataset name

# Display the first few rows
print("First few rows of the dataset:")
print(df.head())

# Step 2: Identify independent and dependent variables
# Based on the dataset, identify 'Construction Cost' as the dependent variable and others as independent variables
dependent_var = 'Construction Cost'  # Adjust this based on your actual dataset
independent_vars = df.columns[df.columns != dependent_var]

print(f"\nIndependent variables: {independent_vars}")
print(f"Dependent variable: {dependent_var}")

# Step 3: Check for missing values
print("\nMissing values in each column:")
missing_values = df.isnull().sum()
print(missing_values)

# Handle missing values (example: fill numerical columns with the mean)
df.fillna(df.mean(), inplace=True)

# Step 4: Generate summary statistics for numerical variables
print("\nSummary statistics for numerical variables:")
summary_stats = df.describe()
print(summary_stats)

# Step 5: Create a correlation heatmap
correlation_matrix = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()

# Interpretation of Correlation Heatmap
# Construction Cost and its correlation with other variables can be interpreted from the heatmap.
# Positive correlations will show closer to 1, and negative correlations closer to -1.
```
